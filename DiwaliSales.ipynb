```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
```

Read Data with insights

```python

df = pd.read_csv("https://github.com/rishabhnmishra/Python_Diwali_Sales_Analysis/raw/main/Diwali%20Sales%20Data.csv",encoding="unicode_escape")
print(df.shape);      #To know dimension
print(df.info());     #Details about data
print(df.head());     #Top 5 rows
df.drop(['Status','unnamed1'],axis=1,inplace=True);   #Delete unnecessary columns
print(df.isnull().sum());    #Check for null values
df.dropna(inplace=True); #Drop rows with null values
print(df.isnull().sum());
print(df);               #Display data
print(df.describe());    #Summary of data
df['Amount']=df['Amount'].astype('int');  #Conversion of datatype
print(df.dtypes);
df.rename(columns={'Marital_Status':'Married'},inplace=True);  #Renaming
print(df.columns);
```

Exploitary Data Analytics

**Insights with Gender**

```python
ax=sns.countplot(x='Gender',data=df,palette='Set2');
for i in ax.containers:
  ax.bar_label(i);
plt.show();
bx=df.groupby(['Gender'])['Amount'].sum().plot(kind='barh',color=['Purple','Gold']);
for i in bx.containers:
  bx.bar_label(i);
plt.show();
```

*Female bought more products in terms of number & money than male*

**Insights with AgeGroup**

```python
ax=sns.countplot(x='Age Group',data=df,palette='Set2',hue='Gender');
for i in ax.containers:
  ax.bar_label(i);
plt.show();
bx=df.groupby(['Age Group'])['Orders'].sum().plot(kind="bar",color=['red','blue','green','yellow','black']);
for i in bx.containers:
  bx.bar_label(i);
plt.show();
df2=df[df['Age Group']=='26-35'];
df2.groupby(['Gender'])['Orders'].sum().plot(kind='pie',autopct='%1.2f%%');

plt.show();
```

*26-35 AGE GROUP has highest number of orders , Females have ordered more*

**Insights with Amount**

```python
ax = sns.barplot(x='Zone',y='Amount',data=df,palette='bright',estimator='sum',errorbar=('ci',0));
for i in ax.containers:
  ax.bar_label(i);
plt.show();
```

*Central Zone has most amount of sales*

```python
df.groupby(['Married'])['Orders'].sum().plot(kind='barh',color=['orange','yellow']);
plt.show();
```

*Females made more orders than males*





**Insights of Occupation**

```python
plt.figure(figsize=(20,8));
bf = df.groupby(['Occupation'])['Orders'].sum().plot(kind='bar',color=["red","green","lime","royalblue"]);
for i in bf.containers:
  bf.bar_label(i);
plt.show();
```


*People working in IT Sector have made most no of orders*




**Insights of Product Category**

```python
plt.figure(figsize=(24,8));
sns.countplot(x='Product_Category',data=df,palette='bright',width=0.6);
plt.show();
plt.figure(figsize=(10,10));
df.groupby(['Product_Category'])['Amount'].sum().plot(kind='pie',autopct='%1.2f%%',radius=2,legend=True,labels=None);
plt.show();
```

*Clothing has most number of orders but food has more amounts*
